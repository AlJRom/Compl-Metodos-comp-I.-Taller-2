```python
def PhiB(t):
    return np.pi * r**2 * B0 * np.cos(Omega*t) * np.cos(2*np.pi*f*t)

# Derivada de la función de flujo magnético
def dPhiB_dt(t):
    return (PhiB(t + 1e-9) - PhiB(t)) / 1e-9

# Ley de Faraday-Lenz para la corriente inducida
def I(t):
    return -dPhiB_dt(t) / R

# Calcular las primeras tres raíces de la función de corriente inducida
root_guesses = [0.1, 0.2, 0.3]  # Ajustar según sea necesario
roots = fsolve(I, root_guesses)

print("Las primeras tres raíces son:", roots)

# Graficar la corriente inducida en función del tiempo
t = np.linspace(0, 1, 1000)  # Ajustar según sea necesario
I_t = I(t)
```
